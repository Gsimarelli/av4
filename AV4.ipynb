## CÃ³digo 1


![alt text](<codigo 1.png>)

```python
import numpy as np
import matplotlib.pyplot as plt

def parte1_2_masses(m1=1.0, k=1.0):
    # Cria um vetor com os valores de m2 de 1 a 10 (inclusive)
    m2_vals = np.arange(1, 10.1, 1)
    frequencias = []  # Lista para armazenar frequÃªncias

    # Loop para cada valor de m2
    for m2 in m2_vals:
        # Matriz de rigidez para 2 massas com conexÃ£o periÃ³dica
        K = k * np.array([[2, -1],
                          [-1, 2]])

        # Matriz de massa diagonal com m1 e m2
        M = np.diag([m1, m2])

        # Calcula A = Mâ»Â¹ K
        A = np.linalg.inv(M) @ K

        # Calcula os autovalores (lambda) e ignora os autovetores
        eigvals, _ = np.linalg.eig(A)

        # FrequÃªncias naturais sÃ£o as raÃ­zes dos autovalores
        omega = np.sqrt(np.abs(eigvals))

        # Ordena para manter a correspondÃªncia entre modos
        omega.sort()

        # Salva o resultado
        frequencias.append(omega)

    # Converte a lista para array numpy
    frequencias = np.array(frequencias)

    # Cria grÃ¡fico com as frequÃªncias vs m2
    plt.figure(figsize=(8,5))
    for i in range(frequencias.shape[1]):
        plt.plot(m2_vals, frequencias[:, i], marker='o', label=f'Modo {i+1}')
    plt.title("FrequÃªncias Naturais vs m2")
    plt.xlabel("m2 (kg)")
    plt.ylabel("FrequÃªncia Natural (rad/s)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Chama a funÃ§Ã£o para rodar
parte1_2_masses()
```

## CÃ³digo 2


![alt text](<CÃ³digo 2.png>)

```python
def parte2_3_masses(m1=1.0, m2=1.0, m3=1.0, k=1.0):
    # Matriz de rigidez para 3 massas com conexÃµes cÃ­clicas
    K = k * np.array([
        [ 2, -1, -1],
        [-1,  2, -1],
        [-1, -1,  2]
    ])

    # Matriz de massa diagonal com m1, m2, m3
    M = np.diag([m1, m2, m3])

    # Calcula A = Mâ»Â¹ K
    A = np.linalg.inv(M) @ K

    # Calcula autovalores e autovetores
    eigvals, eigvecs = np.linalg.eig(A)

    # FrequÃªncias naturais (raiz dos autovalores)
    omega = np.sqrt(np.abs(eigvals))

    # Ordena os resultados para manter ordem nos grÃ¡ficos
    idx = np.argsort(omega)
    omega = omega[idx]
    eigvecs = eigvecs[:, idx]

    # Mostra os resultados no terminal
    print(f"\nConfiguraÃ§Ã£o: m1={m1}, m2={m2}, m3={m3}")
    for i in range(3):
        print(f"Modo {i+1}: Ï = {omega[i]:.4f} rad/s")
        print(f"  Vetor prÃ³prio: {eigvecs[:, i]}")

    # GrÃ¡fico de barras dos modos normais
    fig, axes = plt.subplots(1, 3, figsize=(12, 3))
    labels = ['m1', 'm2', 'm3']
    x = np.arange(3)

    for i in range(3):
        axes[i].bar(x, eigvecs[:, i], color='lightcoral')
        axes[i].set_xticks(x)
        axes[i].set_xticklabels(labels)
        axes[i].set_title(f'Modo {i+1}\nÏ = {omega[i]:.2f}')
        axes[i].axhline(0, color='black', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Exemplos:
parte2_3_masses(1, 1, 1)       # Todas iguais
parte2_3_masses(1, 2, 1)       # Uma distinta
parte2_3_masses(1, 2, 3)       # Todas diferentes
```

```python
def parte3_3_molas(m=1.0, k1=1.0, k2=1.0, k3=1.0):
    # Matriz de rigidez com diferentes constantes de mola
    K = np.array([
        [ k1 + k3, -k1,     -k3],
        [-k1,      k1 + k2, -k2],
        [-k3,     -k2,      k2 + k3]
    ])

    # Matriz de massa com massas iguais
    M = np.diag([m, m, m])

    # Calcula A = Mâ»Â¹ K
    A = np.linalg.inv(M) @ K

    # Calcula autovalores e autovetores
    eigvals, eigvecs = np.linalg.eig(A)

    # FrequÃªncias naturais (raiz dos autovalores)
    omega = np.sqrt(np.abs(eigvals))

    # Ordena os resultados
    idx = np.argsort(omega)
    omega = omega[idx]
    eigvecs = eigvecs[:, idx]

    # Mostra resultados no terminal
    print(f"\nConfiguraÃ§Ã£o de molas: k1={k1}, k2={k2}, k3={k3}")
    for i in range(3):
        print(f"Modo {i+1}: Ï = {omega[i]:.4f} rad/s")
        print(f"  Vetor prÃ³prio: {eigvecs[:, i]}")

    # GrÃ¡ficos dos modos normais
    fig, axes = plt.subplots(1, 3, figsize=(12, 3))
    labels = ['m1', 'm2', 'm3']
    x = np.arange(3)

    for i in range(3):
        axes[i].bar(x, eigvecs[:, i], color='mediumseagreen')
        axes[i].set_xticks(x)
        axes[i].set_xticklabels(labels)
        axes[i].set_title(f'Modo {i+1}\nÏ = {omega[i]:.2f}')
        axes[i].axhline(0, color='black', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Exemplos:
parte3_3_molas(k1=1, k2=2, k3=1)
parte3_3_molas(k1=1, k2=1, k3=3)
```
