## Código 1


![alt text](<codigo 1.png>)

```python
import numpy as np
import matplotlib.pyplot as plt

def parte1_2_masses(m1=1.0, k=1.0):
    # Cria um vetor com os valores de m2 de 1 a 10 (inclusive)
    m2_vals = np.arange(1, 10.1, 1)
    frequencias = []  # Lista para armazenar frequências

    # Loop para cada valor de m2
    for m2 in m2_vals:
        # Matriz de rigidez para 2 massas com conexão periódica
        K = k * np.array([[2, -1],
                          [-1, 2]])

        # Matriz de massa diagonal com m1 e m2
        M = np.diag([m1, m2])

        # Calcula A = M⁻¹ K
        A = np.linalg.inv(M) @ K

        # Calcula os autovalores (lambda) e ignora os autovetores
        eigvals, _ = np.linalg.eig(A)

        # Frequências naturais são as raízes dos autovalores
        omega = np.sqrt(np.abs(eigvals))

        # Ordena para manter a correspondência entre modos
        omega.sort()

        # Salva o resultado
        frequencias.append(omega)

    # Converte a lista para array numpy
    frequencias = np.array(frequencias)

    # Cria gráfico com as frequências vs m2
    plt.figure(figsize=(8,5))
    for i in range(frequencias.shape[1]):
        plt.plot(m2_vals, frequencias[:, i], marker='o', label=f'Modo {i+1}')
    plt.title("Frequências Naturais vs m2")
    plt.xlabel("m2 (kg)")
    plt.ylabel("Frequência Natural (rad/s)")
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Chama a função para rodar
parte1_2_masses()
```

## Código 2


![alt text](<Código 2.png>)

```python
def parte2_3_masses(m1=1.0, m2=1.0, m3=1.0, k=1.0):
    # Matriz de rigidez para 3 massas com conexões cíclicas
    K = k * np.array([
        [ 2, -1, -1],
        [-1,  2, -1],
        [-1, -1,  2]
    ])

    # Matriz de massa diagonal com m1, m2, m3
    M = np.diag([m1, m2, m3])

    # Calcula A = M⁻¹ K
    A = np.linalg.inv(M) @ K

    # Calcula autovalores e autovetores
    eigvals, eigvecs = np.linalg.eig(A)

    # Frequências naturais (raiz dos autovalores)
    omega = np.sqrt(np.abs(eigvals))

    # Ordena os resultados para manter ordem nos gráficos
    idx = np.argsort(omega)
    omega = omega[idx]
    eigvecs = eigvecs[:, idx]

    # Mostra os resultados no terminal
    print(f"\nConfiguração: m1={m1}, m2={m2}, m3={m3}")
    for i in range(3):
        print(f"Modo {i+1}: ω = {omega[i]:.4f} rad/s")
        print(f"  Vetor próprio: {eigvecs[:, i]}")

    # Gráfico de barras dos modos normais
    fig, axes = plt.subplots(1, 3, figsize=(12, 3))
    labels = ['m1', 'm2', 'm3']
    x = np.arange(3)

    for i in range(3):
        axes[i].bar(x, eigvecs[:, i], color='lightcoral')
        axes[i].set_xticks(x)
        axes[i].set_xticklabels(labels)
        axes[i].set_title(f'Modo {i+1}\nω = {omega[i]:.2f}')
        axes[i].axhline(0, color='black', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Exemplos:
parte2_3_masses(1, 1, 1)       # Todas iguais
parte2_3_masses(1, 2, 1)       # Uma distinta
parte2_3_masses(1, 2, 3)       # Todas diferentes
```

## Código 3

```python
def parte3_3_molas(m=1.0, k1=1.0, k2=1.0, k3=1.0):
    # Matriz de rigidez com diferentes constantes de mola
    K = np.array([
        [ k1 + k3, -k1,     -k3],
        [-k1,      k1 + k2, -k2],
        [-k3,     -k2,      k2 + k3]
    ])

    # Matriz de massa com massas iguais
    M = np.diag([m, m, m])

    # Calcula A = M⁻¹ K
    A = np.linalg.inv(M) @ K

    # Calcula autovalores e autovetores
    eigvals, eigvecs = np.linalg.eig(A)

    # Frequências naturais (raiz dos autovalores)
    omega = np.sqrt(np.abs(eigvals))

    # Ordena os resultados
    idx = np.argsort(omega)
    omega = omega[idx]
    eigvecs = eigvecs[:, idx]

    # Mostra resultados no terminal
    print(f"\nConfiguração de molas: k1={k1}, k2={k2}, k3={k3}")
    for i in range(3):
        print(f"Modo {i+1}: ω = {omega[i]:.4f} rad/s")
        print(f"  Vetor próprio: {eigvecs[:, i]}")

    # Gráficos dos modos normais
    fig, axes = plt.subplots(1, 3, figsize=(12, 3))
    labels = ['m1', 'm2', 'm3']
    x = np.arange(3)

    for i in range(3):
        axes[i].bar(x, eigvecs[:, i], color='mediumseagreen')
        axes[i].set_xticks(x)
        axes[i].set_xticklabels(labels)
        axes[i].set_title(f'Modo {i+1}\nω = {omega[i]:.2f}')
        axes[i].axhline(0, color='black', linewidth=0.5)

    plt.tight_layout()
    plt.show()

# Exemplos:
parte3_3_molas(k1=1, k2=2, k3=1)
parte3_3_molas(k1=1, k2=1, k3=3)
```




## 📝 **Relatório — Análise de Frequências Naturais e Modos Normais em Cadeias Atômicas**

Neste estudo, implementamos em Python um modelo numérico para calcular as frequências naturais de vibração e os modos normais de cadeias atômicas lineares com ligações periódicas, abordando sistemas com 2 e 3 átomos por célula unitária. As equações diferenciais do sistema foram reduzidas a um problema de autovalores generalizado, cuja solução fornece as frequências naturais como as raízes quadradas dos autovalores e os modos normais como os autovetores correspondentes.

---

### Parte 1 – Cadeia com 2 átomos

Na primeira parte, consideramos uma cadeia composta por duas massas $m_1 = 1$ kg e $m_2 \in [1, 10]$ kg, conectadas ciclicamente por molas de constante $k = 1$ N/m. Observamos que à medida que $m_2$ aumenta, uma das frequências naturais diminui, enquanto a outra permanece aproximadamente constante. Isso reflete a quebra da simetria no sistema. Quando $m_1 = m_2$, os modos normais correspondem a vibrações simétricas (modo de fase e fora de fase). Com massas desiguais, os modos passam a representar oscilações mais localizadas em uma das massas.

---

### Parte 2 – Cadeia com 3 átomos

Na segunda parte, modelamos uma cadeia com três átomos conectados ciclicamente. Foram analisadas três configurações: massas iguais, uma distinta e todas diferentes. No caso simétrico (massas iguais), obtemos três modos bem definidos: um modo uniforme (todos os átomos se movem em fase), um modo de onda estacionária e um modo antissimétrico. Quando introduzimos assimetrias nas massas, os modos normais se deformam, apresentando amplitudes diferentes para cada massa. Isso demonstra como pequenas alterações nas propriedades inerciais do sistema afetam significativamente os padrões vibracionais.

---

### Parte 3 – Variação das constantes de mola

Por fim, fixamos as massas e variamos as constantes das molas $k_1, k_2, k_3$, mantendo a ligação periódica. Verificamos que a rigidez afeta diretamente as frequências: ligações mais rígidas aumentam a frequência do modo associado àquele elo. Além disso, modos que antes eram simétricos passam a mostrar assimetrias nas amplitudes. Em casos extremos (uma mola muito mais rígida), há modos localizados — onde dois átomos quase não se movem, e a vibração se concentra em apenas um.

---

### Conclusões

As análises mostram que tanto a **simetria da massa** quanto a **distribuição da rigidez** têm impacto direto nas **frequências naturais** e **modos normais**. Sistemas simétricos produzem modos harmônicos bem distribuídos, enquanto perturbações (massas ou molas diferentes) introduzem localizações, alterações de fase e variações significativas nas amplitudes. Isso é fundamental na física de materiais, pois essas características influenciam propriedades como condução de calor e comportamento vibracional de cristais. Os gráficos e modos obtidos numérica e visualmente reforçam a importância das condições de contorno e dos parâmetros físicos na dinâmica vibracional de sistemas discretos periódicos.
